## 4. Create dataframe for modeling

```python
#isolate desired columns
model_df = df[['Score', 'Genres', 'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
               'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature',
               'song_id', 'duration_ms',  'popularity', 'num_markets', 'disc_number', 'is_explicit',
               'track_number', 'album_id']]
```

```python
model_df.head()
```

```python
model_df['album_id'].nunique()
```

```python
model_df.Genres.value_counts()
```

```python
#reduce number of genres by replacing subgenres with main genres
model_df['Genres'].replace(['Pop/R&BRock'], 'Pop/R&B', inplace=True)
model_df['Genres'].replace(['null'], 'Pop/R&B', inplace=True)
model_df['Genres'].replace(['RockExperimental'], 'Experimental', inplace=True)
model_df['Genres'].replace(['ElectronicRockExperimentalGlobal'], 'Experimental', inplace=True)
model_df['Genres'].replace(['Pop/R&BJazz'], 'Pop/R&B', inplace=True)
model_df['Genres'].replace(['RapFolk/Country'], 'Rap', inplace=True)
model_df['Genres'].replace(['ElectronicJazzRock'], 'Electronic', inplace=True)
model_df['Genres'].replace(['ExperimentalPop/R&B'], 'Pop/R&B', inplace=True)
model_df['Genres'].replace(['Pop/R&BFolk/Country'], 'Pop/R&B', inplace=True)
model_df['Genres'].replace(['RockFolk/Country'], 'Rock', inplace=True)
model_df['Genres'].replace(['Pop/R&BExperimental'], 'Pop/R&B', inplace=True)
model_df['Genres'].replace(['ElectronicExperimentalFolk/Country'], 'Electronic', inplace=True)
model_df['Genres'].replace(['ElectronicRap'], 'Rap', inplace=True)
model_df['Genres'].replace(['ElectronicJazzPop/R&B'], 'Electronic', inplace=True)
model_df['Genres'].replace(['ElectronicExperimentalMetal'], 'Electronic', inplace=True)
model_df['Genres'].replace(['RockMetal'], 'Metal', inplace=True)
model_df['Genres'].replace(['RockRap'], 'Rock', inplace=True)
model_df['Genres'].replace(['ElectronicExperimentalRock'], 'Rock', inplace=True)
model_df['Genres'].replace(['ExperimentalRapRock'], 'Rock', inplace=True)
model_df['Genres'].replace(['Pop/R&BElectronic'], 'Pop/R&B', inplace=True)
model_df['Genres'].replace(['ElectronicExperimental'], 'Electronic', inplace=True)
model_df['Genres'].replace(['ElectronicGlobal'], 'Electronic', inplace=True)
model_df['Genres'].replace(['ExperimentalElectronic'], 'Electronic', inplace=True)
model_df['Genres'].replace(['ExperimentalJazz'], 'Jazz', inplace=True)
model_df['Genres'].replace(['ExperimentalRock'], 'Experimental', inplace=True)
model_df['Genres'].replace(['ExperimentalMetal'], 'Metal', inplace=True)
model_df['Genres'].replace(['ElectronicJazz'], 'Electronic', inplace=True)
model_df['Genres'].replace(['ElectronicMetal'], 'Electronic', inplace=True)
model_df['Genres'].replace(['ElectronicRock'], 'Rock', inplace=True)
model_df['Genres'].replace(['RapRock'], 'Rap', inplace=True)
model_df['Genres'].replace(['ExperimentalRap'], 'Rap', inplace=True)
model_df['Genres'].replace(['GlobalRap'], 'Rap', inplace=True)
model_df['Genres'].replace(['GlobalRock'], 'Rock', inplace=True)
model_df['Genres'].replace(['Folk/CountryPop/R&B'], 'Folk/Country', inplace=True)
model_df['Genres'].replace(['ElectronicFolk/Country'], 'Folk/Country', inplace=True)
model_df['Genres'].replace(['Folk/CountryPop/R&BRock'], 'Folk/Country', inplace=True)
model_df['Genres'].replace(['GlobalPop/R&B'], 'Pop/R&B', inplace=True)
model_df['Genres'].replace(['MetalRock'], 'Metal', inplace=True)
model_df['Genres'].replace(['JazzMetal'], 'Metal', inplace=True)
model_df['Genres'].replace(['Pop/R&BRap'], 'Pop/R&B', inplace=True)
model_df['Genres'].replace(['JazzPop/R&B'], 'Jazz', inplace=True)
model_df['Genres'].replace(['ElectronicPop/R&B'], 'Pop/R&B', inplace=True)
```

```python
#change is_explicit column bool to numeric
model_df['is_explicit'].replace(to_replace='True', value = 1, inplace=True)
model_df['is_explicit'].replace(to_replace='False', value = 0, inplace=True)
```

```python
#change score column from string to float
model_df['score'] = model_df['Score'].astype(float)
model_df['popularity'] = model_df['popularity'].astype(float)
model_df['mode'] = model_df['mode'].astype(float)
model_df['key'] = model_df['key'].astype(float)
```

```python
#group songs by album_id
grouped_albums = model_df.groupby('album_id')
```

```python
grouped_albums.size()
```

```python
#get mean values of numerical variables
grouped_mean_df = grouped_albums[
    'energy', 'speechiness', 'danceability', 'acousticness', 'instrumentalness', 'loudness', 
    'liveness', 'valence'].agg([np.mean])
```

```python
grouped_mean_df.columns = ["_".join(x) for x in grouped_mean_df.columns.ravel()]
```

```python
grouped_mean_df.head()
```

```python
#get max value of other numerical variables
grouped_max_df = grouped_albums[
    'energy', 'liveness', 'speechiness', 'valence', 'tempo', 'time_signature', 'popularity', 
    'score', 'num_markets', 'is_explicit', 'mode', 'key', 'instrumentalness', 'acousticness',
    'loudness', 'danceability'].agg([np.max])
```

```python
grouped_max_df.columns = ["_".join(x) for x in grouped_max_df.columns.ravel()]
```

```python
grouped_max_df.head()
```

```python
#get album length in time and number of tracks in an album
album_std_duration_df = grouped_albums.agg(album_run_time_ms = pd.NamedAgg(column='duration_ms', aggfunc=np.sum),
                                        number_of_tracks = pd.NamedAgg(column='track_number', aggfunc='count'))
```

```python
album_std_duration_df = album_std_duration_df[album_std_duration_df.number_of_tracks <= 35]
```

```python
album_std_duration_df.head()
```

```python
#concatenate the two dataframes
album_df = pd.concat([album_std_duration_df, grouped_mean_df, grouped_max_df], axis=1, sort=False)
```

```python
#add genres column to dataframe
genres = grouped_albums['Genres'].first()
album_df['Genres'] = genres
```

```python
album_df.Genres.value_counts()
```

```python
#save df for modeling
with open('data/album_df.pkl', 'wb') as f:
    pickle.dump(album_df, f)
```
