```python
#import packages
import pandas as pd
from tqdm import tqdm
import pickle
import numpy as np
```

## 1. Import data from Spotify and Pitchfork

```python
pitchfork_df = pd.read_csv('data/pitchfork_df.csv')
features_df = pd.read_pickle('data/song_features_df.pkl')
objects_df = pd.read_pickle('data/song_objects_df.pkl')
```

```python
pitchfork_df.head()
```

```python
features_df.head()
```

```python
objects_df.head()
```

## 2. Merge Spotify objects and features dataframes

```python
df_spotify = pd.merge(features_df, objects_df, on='song_id')
```

```python
#in order to see all the columns
pd.set_option('display.max_columns', None)
```

```python
df_spotify.head()
```

```python
df_spotify.info()
```

```python
#remove albums with no information
df_spotify = df_spotify[df_spotify['energy'].notna()]
```

```python
df_spotify.head()
```

```python
df_spotify.info()
```

## 3. Merge Spotify and Pitchfork info

```python
#apply lower case to albums and titles for merging
pitchfork_df['title'] = pitchfork_df['title'].astype(str).apply(lambda x: x.lower())
pitchfork_df['artist'] = pitchfork_df['artist'].astype(str).apply(lambda x: x.lower())
```

```python
pitchfork_df.head()
```

```python
#rename columns to match for merging
df_spotify.rename(columns={"Album": "title", "Artist": "artist"}, inplace=True)
```

```python
df_spotify.head()
```

```python
#merge spotify and pitchfork dataframes
df = pitchfork_df.merge(df_spotify, how='inner')
```

```python
#dropping rows with value == null
df = df[(df != 'null').all(axis=1)]
```

```python
df.head()
```

```python
df['album_id'].nunique()
```

## 4. Create dataframe for modeling

```python
#isolate desired columns
model_df = df[['score','genre', 'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
               'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature',
               'song_id', 'duration_ms',  'popularity', 'num_markets', 'disc_number', 'is_explicit',
               'track_number', 'album_id']]
```

```python
model_df.head()
```

```python
model_df.genre.value_counts()
```

```python
model_df.genre.nunique()
```

```python
#change is_explicit column bool to numeric
model_df['is_explicit'].replace(to_replace='True', value = 1, inplace=True)
model_df['is_explicit'].replace(to_replace='False', value = 0, inplace=True)
```

```python
model_df.is_explicit.head()
```

```python
type(model_df.score[0])
```

```python
#change score column from string to float
model_df['score'] = model_df['score'].astype(float)
model_df['popularity'] = model_df['popularity'].astype(float)
model_df['mode'] = model_df['mode'].astype(float)
model_df['key'] = model_df['key'].astype(float)
```

```python
model_df.head()
```

```python
#group songs by album_id
grouped_albums = model_df.groupby('album_id')
```

```python
grouped_albums.size()
```

```python
#get mean values of numerical variables
grouped_mean_df = grouped_albums[
    'energy', 'speechiness', 'danceability', 'acousticness', 'instrumentalness', 'loudness',
    'tempo', 'liveness', 'valence'].agg([np.mean])
```

```python
grouped_mean_df.columns = ["_".join(x) for x in grouped_mean_df.columns.ravel()]
```

```python
grouped_mean_df.head()
```

```python
#get max value of other numerical variables
grouped_max_df = grouped_albums[
    'energy', 'liveness', 'speechiness', 'valence', 'tempo', 'time_signature', 'popularity', 
    'score', 'num_markets', 'is_explicit', 'mode', 'key', 'instrumentalness', 'acousticness',
    'loudness', 'danceability'].agg([np.max])
```

```python
grouped_max_df.columns = ["_".join(x) for x in grouped_max_df.columns.ravel()]
```

```python
grouped_max_df.head()
```

```python
#get album length in time and number of tracks in an album
album_std_duration_df = grouped_albums.agg(album_run_time_ms = pd.NamedAgg(column='duration_ms', aggfunc=np.sum),
                                        number_of_tracks = pd.NamedAgg(column='track_number', aggfunc='count'))
```

```python
album_std_duration_df = album_std_duration_df[album_std_duration_df.number_of_tracks <= 35]
```

```python
album_std_duration_df.head()
```

```python
#concatenate the two dataframes
album_df = pd.concat([album_std_duration_df, grouped_mean_df, grouped_max_df], axis=1, sort=False)
```

```python
#add genres column to dataframe
genres = grouped_albums['genre'].first()
album_df['genre'] = genres
```

```python
album_df.genre.value_counts()
```

```python
album_df.head()
```

```python
album_df = album_df[album_df['number_of_tracks'].notna()]
```

```python
album_df.info()
```

```python
#save df for modeling
with open('data/album_df.pkl', 'wb') as f:
    pickle.dump(album_df, f)
```
